```python
#Reduced dataset for increased speed
df_small = df.head(50000)
```

```python
# Check total missing values per column
print(df_small.isnull().sum())

# Percentage of missing values
print((df_small.isnull().mean() * 100).round(2))
```

```python
# Drop incomplete ID columns
df_small = df_small.drop(columns=[
    'sha', 'doi', 'pmcid', 'pubmed_id', 
    'mag_id', 'who_covidence_id', 'arxiv_id', 
    'pdf_json_files', 'pmc_json_files', 'url', 's2_id'
])

# Fill missing journal/authors with "Unknown"
df_small['journal'] = df_small['journal'].fillna('Unknown')
df_small['authors'] = df_small['authors'].fillna('Unknown')
```

```python
# keep abstracts even if missing 
df_small['abstract'] = df_small['abstract'].fillna('')
```

```python
#convert publish_time to datetime, extract year
df_small['publish_time'] = pd.to_datetime(df_small['publish_time'], errors='coerce')
```

```python
# drop rows where publish_time conversion failed
df_small = df_small.dropna(subset=['publish_time']).reset_index(drop=True)
df_small['year'] = df_small['publish_time'].dt.year.astype(int)
```

```python
# Create abstract_word_count
df_small['abstract_word_count'] = df_small['abstract'].apply(lambda x: len(str(x).split()))
# Create title_word_count 
df_small['title_word_count'] = df_small['title'].apply(lambda x: len(str(x).split()))
```

```python
# Save cleaned dataset (optional)
df_small.to_csv('metadata_cleaned_sample.csv', index=False)
```

```python
#Count papers by publication year
import matplotlib.pyplot as plt

# Count per year (sorted)
year_counts = df_small['year'].value_counts().sort_index()

# Simple bar plot
plt.figure(figsize=(10,4))
plt.bar(year_counts.index.astype(int), year_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of papers')
plt.title('Publications by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#most frequent words in tittles
# Simple text cleaning and counting
import re
from collections import Counter
def clean_text(s):
    s = str(s).lower()
    s = re.sub(r'[^a-z0-9\s]', ' ', s)  # remove punctuation
    s = re.sub(r'\s+', ' ', s).strip()
    return s

titles = df_small['title'].dropna().astype(str).apply(clean_text).tolist()
all_words = " ".join(titles).split()

# use a small stopword list (expand if you want)
stopwords = {
    'the','and','of','in','a','to','for','with','on','by','an','from','is','are','using',
    'study','studies','case','cases','report','reports','based','analysis','effects'
}

filtered = [w for w in all_words if w not in stopwords and len(w)>2]
freq = Counter(filtered)
top_words = freq.most_common(30)

# show as dataframe
import pandas as pd
pd.DataFrame(top_words, columns=['word','count']).head(20)
```

```python
#bar chart representing top most frequent words
import matplotlib.pyplot as plt

# Take top 20 words
top_words = freq.most_common(20)

words = [w for w, _ in top_words]
counts = [c for _, c in top_words]
plt.figure(figsize=(12,6))
plt.bar(words, counts, color="skyblue")
plt.xticks(rotation=45, ha="right")
plt.title("Top 20 Most Frequent Words in Paper Titles")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
# median abstract length by year
abstract_by_year = df_small.groupby('year')['abstract_word_count'].median().dropna()
abstract_by_year.plot(title='Median Abstract Word Count by Year', figsize=(10,4))
plt.xlabel('Year'); plt.ylabel('Median abstract word count'); plt.show()

# Top journals per recent year (filter)
recent = df_small[df_small['year'] >= 2020]
recent['journal'].value_counts().head(10)
```

```python
#abstrct length distribution
plt.figure(figsize=(10,6))
df_small['abstract_word_count'].hist(bins=50, color='orange')
plt.title("Distribution of Abstract Word Counts")
plt.xlabel("Word Count")
plt.ylabel("Number of Papers")
plt.show()
```

```python
#publication trend(yearly growth) to show how COVID-19 research exploded.

yearly_counts = df_small['year'].value_counts().sort_index()
cumulative = yearly_counts.cumsum()

plt.figure(figsize=(12,6))
plt.plot(cumulative.index, cumulative.values, marker='o')
plt.title("Cumulative Publications Over Time")
plt.xlabel("Year")
plt.ylabel("Cumulative Count")
plt.grid(True)
plt.show()
```

```python
#Authors analysis, who are the most prolific authors?
all_authors = df_small['authors'].dropna().astype(str).str.split(';')
flat_authors = [a.strip() for sublist in all_authors for a in sublist if a.strip() != ""]
author_counts = Counter(flat_authors).most_common(10)

print(author_counts)
```
